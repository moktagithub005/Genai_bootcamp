```python
from langchain_text_splitters import HTMLHeaderTextSplitter, RecursiveCharacterTextSplitter
import requests
```

```python
url="https://python.langchain.com/docs/introduction/"
html=requests.get(url,timeout=20).text

headres_to_split=[
    ("h1","header1"),
    ("h2","header2"),
    ("h3","header3"),
    ("h4","header4"),

]

header_splitter=HTMLHeaderTextSplitter(headers_to_split_on=headres_to_split)
header_docs=header_splitter.split_text(html)

final_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_docs=final_splitter.split_documents(header_docs)

print("header-level_chunks")

len(header_docs)
```
